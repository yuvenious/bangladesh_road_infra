```python
import pandas as pd
import numpy as np

from math import radians, cos, sin, sqrt, atan2

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# This function defines the distance between two locations.
def distance(loc1, loc2, R = 6380):
    #change the value into radians
    lat1 = radians(float(loc1[0]))
    lon1 = radians(float(loc1[1]))
    
    lat2 = radians(float(loc2[0]))
    lon2 = radians(float(loc2[1]))
    
    #calculate the distance
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = (sin(dlat/2))**2 + cos(lat1) * cos(lat2) * (sin(dlon/2))**2
    c = 2 * atan2( sqrt(a), sqrt(1-a) )
    d = R * c # where R is the radius of the Earth
    return d
```

## 2. Convert road/bridge data into the form readable in Simio

```python
# define a couple of functions in order to manipulate data frame later
# to be used in lambda function

def obj_class(x):
    """
    Translate Names in road/bridge data into,
    Actual Name in Simio models ("BasicNode", "TransferNode", "Server")
    """
    if x == "Road":
        return "BasicNode"
    elif x == "Output@Bridge":
        return "TransferNode"
    elif x == "Bridge":
        return "Bridge"
    
def input_trans(x):
    """
    If there is a bridge after the road, create "Input@BridgeXXX" object.
    """
    if x[:6] == "Bridge":
        return "".join(["Input@", x])
    else:
        return x
```

### 1) Define function to extract relevant data

```python
def create_data(df_r, df_b, nrow = False):
    """
    Extract necessary information from road/bridge data
    1. Type (Bridge/Road), LRP
    2. Longitude, Latitude, Chainage, Distance
    3. Bridge condition(A/B/C/D) and length
    """
    road = pd.concat([df_r[["lon", "lat", "chainage", "lrp"]],
                      pd.DataFrame(("Road "*len(df_r)).split(), columns = ["type"])], axis = 1)

    bridge = df_b.loc[:, ["lon", "lat", "chainage", "LRPName", "length", "condition", "length"]]
    bridge.columns = ['lon', 'lat', 'chainage', 'lrp', 'BriLen', 'BriCond', 'type']
    bridge.loc[:, "type"] = ("Bridge "*len(df_b)).split()
    transfernode = bridge[:]
    transfernode.loc[:, "type"] = "Output@Bridge"
    transfernode.loc[:, "BriLen"] = None

    df = pd.concat([road, bridge, transfernode]).sort_values(by = ["chainage", "type"])
    df.reset_index(inplace = True)
    df["Object Name"] = df.loc[:, ["index", "type"]].apply(lambda x: x["type"] + str(x["index"]), axis = 1)
    
    df["distance"] = [1e3 * distance(df[["lat", "lon"]].iloc[i],
                                     df[["lat", "lon"]].iloc[i+1]) for i in range(len(df) - 1)] + [None]
    if type(nrow) == int:
        df = df.loc[:nrow]
    elif not nrow:
        df = df
    return df
data = create_data(df_r, df_b, 5)
data
```

### 2) Convert into Link sheet

```python
def convert_link(data):
    col = "Link Class;Link Name;From Node;To Node;Network;Type;DrawnToScale;LogicalLength".split(";")
    sheet = pd.DataFrame(columns = col)
    sheet["Link Class"] = ("Path " * (len(data) - 1)).split()
    sheet["Type"] = ("Unidirectional " * (len(data) - 1)).split()
    sheet["Network"] = ("NetworkA " * (len(data) - 1)).split()
    sheet["DrawnToScale"] = ("False " * (len(data) - 1)).split()
    sheet["From Node"] = data["Object Name"][:-1]
    sheet["To Node"] = data["Object Name"][1:].reset_index(drop = True)
    sheet["To Node"] = sheet["To Node"].apply(lambda x: input_trans(x))
    sheet["num"] = list(range(1, len(data)))
    sheet["Link Name"] = sheet.apply(lambda x: x.loc["Link Class"] + str(x.loc["num"]), axis = 1)
    sheet["LogicalLength"] = data["distance"]
    sheet.drop("num", axis = 1, inplace = True)
    idx = sheet["From Node"].apply(lambda x:x[:6]) != "Bridge"
    return sheet[idx]
convert_link(data)
```

### 3) Convert into Object sheet

```python
def convert_obj(data, multi = 100, ng = -1):
    # create spread sheet for "Object"
    col = ['Object Class', 'Object Name', 'X', 'Y', 'Z', 'Length', 'Width', 'Height',
           'BriLen', 'BriCond',
          ]
    
    sheet = pd.DataFrame(columns = col)
    sheet["Object Class"] = data["type"].apply(lambda x: obj_class(x))
    sheet["Object Name"] = data["Object Name"]
    sheet["X"] = data["lon"].apply(lambda x: x * multi)
    sheet["Y"] = np.repeat(0, len(data))
    sheet["Z"] = data["lat"].apply(lambda x: x * multi * ng)
    sheet["BriLen"] = data["BriLen"]
    sheet["BriCond"] = data["BriCond"]
    sheet["chainage"] = data["chainage"]
    
    sheet = sheet[sheet["Object Class"] != "TransferNode"]
    return sheet
convert_obj(data).tail()
```
